```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.array([1, 8])
y = np.array([3, 10])
plt.plot(x, y,color='Black')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.figure(figsize=(8,6))
plt.plot(x,y,color='Grey',linestyle=':',marker='s',linewidth=5)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line Plot')
plt.legend(['my data series'])
plt.grid(True)
plt.show()
```

```python
categories=['A','B','C','D','E']
values=[5,2,3,4,6]
plt.figure(figsize=(8,5))
plt.bar(categories,values,color='Grey',edgecolor='Black')
plt.xlabel('categories')
plt.ylabel('values')
plt.title('Bar chart')
plt.show()
```

```python
data=[12,15,13,17,16,16,19,12,14,16,14,17,18,18,15,14]
plt.hist(data,bins=16,edgecolor='Black',color='Grey')
plt.xlabel('Value')
plt.ylabel('frequency')
plt.title('Simple Histogram Example')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [5, 7, 4, 6, 8]
plt.scatter(x, y,color='Grey',edgecolors='Black')
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.title('Simple Scatter Plot Example')
plt.show()
```

```python
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['#999999', '#BBBBBB', '#DDDDDD', '#EEEEEE']
plt.pie(sizes,
        labels=labels,
        colors=colors,
        autopct='%1.1f%%',
        startangle=140)
plt.title('Simple Pie Chart Example')
plt.axis('equal')
plt.show()
```

```python
data = [7, 8, 5, 6, 7, 9, 12, 5, 7, 8, 6, 12, 2, 6]
plt.boxplot(data)
plt.title('Simple Box Plot Example')
plt.ylabel('Values')
plt.show()
```

```python
data = np.random.rand(10, 10)
plt.imshow(data, cmap='Greys', interpolation='nearest')
plt.colorbar()
plt.title('Simple Heatmap Example')
plt.show()
```
